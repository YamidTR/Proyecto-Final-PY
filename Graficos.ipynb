```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
df = pd.read_csv('DatosDeColombia.csv', encoding = 'utf8')
```

```python
df
```

```python
plt.bar(df["DEPARTAMENTO"], df["TASA_MATRICULACIÓN_5_16"], color= "red")
plt.title("Gráfico de Barras")
plt.xticks(rotation=60)
plt.xlabel("DEPARTAMENTO")
plt.ylabel("TASA DE MATRICULACION")
plt.show()
```

```python
plt.fill_between(df["AÑO"], df["REPROBACIÓN"], color= "blueviolet")
plt.title("Gráfico de Barras")
plt.xticks(rotation=0)
plt.xlabel("AÑO")
plt.ylabel("REPROBACIÓN")
plt.show()
```

```python
plt.plot(df["AÑO"], df["REPITENCIA"], color= "brown")
plt.title("Gráfico de Barras")
plt.xticks(rotation=0)
plt.xlabel("AÑO")
plt.ylabel("REPITENCIA")
plt.show()
```
